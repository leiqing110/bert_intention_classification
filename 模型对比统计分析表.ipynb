```python
import pandas as pd
```

```python
# id获取类别名称
class_list = pd.read_csv('data/custom/labels.txt',sep='\t',dtype=str,names=['label']).label.to_list()
id2label = {str(id):value for id ,value in enumerate(class_list)}
id2label
```

```python
output_cols = ['query','label_idx_gt','label_val_gt','diff_val','entity','model1_pred_idx','model1_pred_val','model2_pred_idx','model2_pred_val']
test_data = pd.read_csv('data/custom/test_data.txt',sep='\t',dtype=str,names = ['query','label_idx_gt'])
test_data['label_val_gt'] = test_data['label_idx_gt'].map(id2label)
test_data
```

```python
model1_df = pd.read_csv('test_res_entity0816.txt',sep='\t',dtype=str)
model1_df.rename(columns={'pred':'model1_pred_idx'},inplace=True)
model1_df['model1_pred_val'] = model1_df['model1_pred_idx'].map(id2label)
model1_df
```

```python
model2_df = pd.read_csv('test_res_entity0818_.txt',sep='\t',dtype=str)
model2_df.rename(columns={'pred':'model2_pred_idx'},inplace=True)
model2_df['model2_pred_val'] = model2_df['model2_pred_idx'].map(id2label)
model2_df
```

```python
test_data[['model1_pred_idx','model1_pred_val']] = model1_df[['model1_pred_idx','model1_pred_val']]
test_data[['model2_pred_idx','model2_pred_val']] = model2_df[['model2_pred_idx','model2_pred_val']]
test_data
```

```python
# 定义diff_val，
# 0表示model1和model2预测都正确
# 1表示model1预测正确
# 2表示model2预测正确
# 3表示model1和model2都识别错误
def diff_val(model1,model2,label_idx):
    if model1 == model2 and model1 == label_idx:
        return '0'
    elif model1 != model2 and model1 == label_idx:
        return '1'
    elif model1 != model2 and model2 == label_idx:
        return '2'
    else:
        return '3'
test_data['diff_val'] = test_data.apply(lambda row :diff_val(row['model1_pred_idx'],row['model2_pred_idx'],row['label_idx_gt']),axis =1)
test_data
```

```python
# 四大类表示
# 0表示其他
# 1表示品牌
# 2表示厂商
# 3表示车系
# 4表示车型
def entity(label_idx):
    all_entity = ['7','8','9','14','15']
    if label_idx not in all_entity:
        return '0'
    elif label_idx == '7':
        return '1'
    elif label_idx == '8':
        return '2'
    elif label_idx == '9':
        return '3'
    elif label_idx in ['14','15']:
        return '4'

test_data['entity'] = test_data.apply(lambda row :entity(row['label_idx_gt']),axis =1)
test_data
```

```python
test_data.to_csv('compare_test.csv',index=None,sep=',',encoding='utf-8-sig',columns=output_cols)
```
